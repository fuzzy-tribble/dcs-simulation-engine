# Template for Accessing and Evaluating Simulation Data

```python
import dcs_simulation_engine.helpers.database_helpers as dbh

# also import matplotlib, sns, pandas, and whatever other libraries you need for analysis/visualization
```

```python
db = dbh.get_db()
db.name
```

```python
# Access the players and runs collection
players = list(db["players"].find())
runs = list(db["runs"].find())
```

```python
# Count players, list by _id, etc.
print(f"Total players in DB: {len(players)}")
print("Player IDs:")
for p in players:
    print(f" - {p['_id']}")
```

```python
# How many runs are in the database?
print(f"Total runs in DB: {len(runs)}")

# What keys are available in a run document?
if runs:
    print("Keys in a run document:")
    for key in runs[0].keys():
        print(f" - {key}")
```

```python
# What games have been run/played?
for r in runs:
    print(
        f"Run ID: {r['_id']}, Game: {r['game_config']['name']}, Player: {r['player_id']}"
    )
```
